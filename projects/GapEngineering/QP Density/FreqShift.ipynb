This is from Martinis 2011 PRB measurement of energy decay in superconducting
qubits from nonequilibrium quasiparticles
from equation (3)
$\frac{Y_{j}(\omega)}{1/R_{n}}=(1+i)\sqrt{2}(\frac{\Delta}{\hbar \omega})^{3/2}
\frac{n_{qp}}{n_{cp}}-i\pi\frac{\Delta}{\hbar \omega}$
In order to get the resonance condition, the imaginary part will get balanced 
from the capacitance of the junction.
$Im[Y_{j}+i\omega C ] = 0$, thus, with $x_{qp}=\frac{n_{qp}}{n_{cp}}$ we get the balance condition
$\sqrt{2}(\frac{\Delta}{\hbar \omega})^{3/2}
x_{qp}-\pi\frac{\Delta}{\hbar \omega} + \omega CR_{n} = 0$, or in frequency 
$\sqrt{2}(\frac{\Delta}{h f})^{3/2}
x_{qp}-\pi\frac{\Delta}{h f} + 2\pi f CR_{n} = 0$
also, the energy gap is
$\Delta = \Delta_{0}(1-x_{qp})$
we can first take the gap as constant and as dependent, let's see.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import *
def getQPdensity(f):
    delta = 190*elementary_charge*10**(-6)
    C = 50.97*10**(-15)
    A = delta/(h*f)
    R_n = 16792.67
    x_qp = (pi*A-2*pi*f*C*R_n)/((2)**(1/2)*A**(3/2))
    f0 = 5.1805*10**9
    print((delta/(2*h*f0**2))/C)
    return x_qp
```

```python
f_list = [5.1805e9,5.180e9, 5.1775e9, 5.1745e9, 5.171e9, 5.1645e9, 5.156e9]
QP_list = []
for f in f_list:
    QP_list.append(getQPdensity(f))

power_list = [-10, -8, -6, -4, -2, 0]
# power_list = [10**((p-45)/10) for p in power_list]
```
